```python
import pandas as pd
import numpy as np
df = pd.read_csv('titanic.csv')
df.info()
```

```python
df = df.dropna()
df
```

```python
df.info()
```

```python
cols = ['Parch','Fare']
df = df.drop(cols,axis=1)
df
```

```python
df.info()
```

```python
dummies = []
cols = ['Pclass','Embarked']
for col in cols:
  dummies.append(pd.get_dummies(df[col]))
titanic_dummies = pd.concat(dummies , axis=1)
df = pd.concat((df,titanic_dummies),axis = 1)
df
```

```python
df.info()
```

```python
cols = ['C']
df = df.drop(cols,axis=1)
df
```
